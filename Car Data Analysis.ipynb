```python
def Univariate(dataset,quan):
    descriptive=pd.DataFrame(index=["Mean","Median","Mode","Min","25%","50%","75%","100%","Max","IQR",
                                   "1.5IQR","LesserRange","GreaterRange","skew","kurtosis","var","std"],columns=quan)
    for columnName in quan:
        descriptive[columnName]["Mean"]=dataset[columnName].mean()
        descriptive[columnName]["Median"]=dataset[columnName].median()
        descriptive[columnName]["Mode"]=dataset[columnName].mode()[0]
        descriptive[columnName]["Min"]=dataset[columnName].min()
        descriptive[columnName]["25%"]=dataset.describe()[columnName]["25%"]
        descriptive[columnName]["50%"]=dataset.describe()[columnName]["50%"]
        descriptive[columnName]["75%"]=dataset.describe()[columnName]["75%"]
        descriptive[columnName]["100%"]=dataset.describe()[columnName]["max"]
        descriptive[columnName]["Max"]=dataset[columnName].max()
        descriptive[columnName]["IQR"]=descriptive[columnName]["75%"]-descriptive[columnName]["25%"]
        descriptive[columnName]["1.5IQR"]=1.5*descriptive[columnName]["IQR"]
        descriptive[columnName]["LesserRange"]=descriptive[columnName]["25%"]- descriptive[columnName]["1.5IQR"]
        descriptive[columnName]["GreaterRange"]=descriptive[columnName]["75%"]+ descriptive[columnName]["1.5IQR"]
        descriptive[columnName]["skew"]=dataset[columnName].skew()
        descriptive[columnName]["kurtosis"]=dataset[columnName].kurtosis()  
        descriptive[columnName]["var"]=dataset[columnName].var() 
        descriptive[columnName]["std"]=dataset[columnName].std() 
    return descriptive
```

Normal Distribution(Probability Dencity Function)

```python
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # generate a sample
    sample = dataset
    # calculate parameters
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    # define the distribution
    dist = norm(sample_mean, sample_std)
    
    # sample probabilities for a range of outcomes
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
    
```

```python
get_pdf_probability(dataset['Price'],1000000,3000000)
```

Z- Score: Normal distribution to Standard Normal Distribution

```python
def stdNBgraph(dataset):
        # Coverted to standard Normal Distribution
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()

    values=[i for i in dataset]

    z_score=[((j-mean)/std) for j in values]

    sns.distplot(z_score,kde=True)

    sum(z_score)/len(z_score)
    #z_score.std()
```

Varience Inflation Factor

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(x):
    vif=pd.DataFrame()
    vif["variables"]=x.columns
    vif["VIF"]=[variance_inflation_factor(x.values,i)for i in range(x.shape[1])]
    return(vif)
```

T-Test

Independant Sample
Diferrent Group(Honda, Mahindra) but same contion(Price)

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
Honda= dataset[dataset['Make']=='Honda']['Price']
Mahindra = dataset[dataset['Make']=='Mahindra']['Price']
ttest_ind(Honda, Mahindra)
```

p>0.05% , 1.05>0.05% 
H0=(Null Hypothesis)There is no significant difference between the two columns H1=(Alternative Hypothesis)There is significant difference between the two columns.
So accept the H0=Null hypothesis(There is no significant difference between the two columns). Reject the H1=Alternative Hypothesis(there is significant difference between the two columns).

Dependant Sample
Same Group(Male) but Different Condition(ssc_p,hsc_p)

```python
from scipy.stats import ttest_rel
Honda = dataset[dataset['Make']=='Honda']['Price']
Honda1 = dataset[dataset['Make']=='Honda']['Year']
ttest_rel(Honda, Honda1)
```

p>0.05% , 9.39>0.05%
Accept the H0=Null hypothesis(There is no significant difference between the two columns). 
Reject the H1=Alternative Hypothesis(There is significant difference between the two columns).
